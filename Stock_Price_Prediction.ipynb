```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
from subprocess import check_output
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from pandas.plotting import lag_plot
from pandas import datetime

from sklearn.metrics import mean_squared_error
warnings.filterwarnings('ignore')
```

```python
print(os.listdir(r"E:\PROJECTS\Stock-Price-Prediction\Stock Price Prediction"))

```

```python
df = pd.read_csv("E:\PROJECTS\Stock-Price-Prediction\Stock Price Prediction\Dataset\AAPL.csv")
df.head()
```

```python
print(df.head())
print(df.shape)
print(df.columns)
```

```python
df[['Close']].plot()
plt.title("APPLE")
plt.show()
```

```python
# Comulative Return
dr = df.cumsum()
dr.plot()
plt.title('apple Cumulative Returns')
```

```python
train_data, test_data = df[0:int(len(df)*0.8)], df[int(len(df)*0.8):]
plt.figure(figsize=(12,7))
plt.title('apple Prices')
plt.xlabel('Dates')
plt.ylabel('Prices')
plt.plot(df['Open'], 'blue', label='Training Data')
plt.plot(test_data['Open'], 'green', label='Testing Data')
#plt.xticks(np.arange(0,1857, 300), df['Date'][0:1857:300])
plt.legend()
```
